# Scraped Tweet Data Exploring
This Notebook was used to generate graphics and explore the difference the twitter datasets before and after classification.
The key word search for filtering by disaster type is activated here.

This notebook also generates the list of valid countries to querry in the map.




```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
code_dir = os.getcwd()
parent_dir = os.path.dirname(code_dir)
data_dir = os.path.join(parent_dir,"Data")
tweet_dir = os.path.join(parent_dir,"tweets_archive")
```

```python
wxevents= ["tornado","blizzard","snow", "ice storm","sleet","thunderstorm","hail","freeze","frost","extreme cold","lightning",    "winds"       
"gale","storm"]
movement=["landslide" , "tsunami" , "tidal wave" , "earthquake" , "eruption" , 
          "volcano" , "lava" , "lahar" , "avalanche" , "mudslide" , "sinkhole"]
troplist=["typhoon" ,"cyclone","hurricane", "tropical storm","storm surge"]
flist= ["fire", "forest fire", "wildfire", "bushfire", "extreme heat", "record heat", "heatwave" , "heat wave" ]
floods = ["flood","flooded","flooding"]
```

### Put all files into a list of dataframes
Lists of datframes make concatenating easy later on

```python
### dfs will hold the raw scraped tweets
dfs=[]
for table in os.listdir(os.path.join(tweet_dir)):
    if table.startswith("proc"):
        tempdf = pd.read_csv(os.path.join(tweet_dir,table), engine = 'python', )
        tempdf['isDate'] = tempdf.Datetime.str.startswith('2')
        tempdf = tempdf[tempdf['isDate'] == True]
        try:
            tempdf = tempdf.drop(columns = ['TweetLink','isDate'])
        except:
            tempdf = tempdf
        dfs.append(tempdf)
        
### dfs2 will hold the scraped tweets after classification
dfs2=[]
for table in os.listdir(os.path.join(tweet_dir)):
    if table.startswith("result"):
        tempdf = pd.read_csv(os.path.join(tweet_dir,table), engine = 'python', )
        tempdf['isDate'] = tempdf.Datetime.str.startswith('2')
        tempdf = tempdf[tempdf['isDate'] == True]
        try:
            tempdf = tempdf.drop(columns = ['TweetLink','isDate'])
        except:
            tempdf = tempdf.drop(columns = ['isDate'])
        dfs2.append(tempdf)
        
### dfs3 will hold the scraped tweets after geoing
tweet_dir = os.path.join(parent_dir,"GeoTweets")
dfs3=[]
for table in os.listdir(os.path.join(tweet_dir)):
    if table.startswith("geo"):
        tempdf = pd.read_csv(os.path.join(tweet_dir,table), engine = 'python', )
        tempdf['isDate'] = tempdf.Datetime.str.startswith('2')
        tempdf = tempdf[tempdf['isDate'] == True]
        try:
            tempdf = tempdf.drop(columns = ['TweetLink','0','isDate'])
        except:
            tempdf = tempdf
        dfs3.append(tempdf)
```

### Concat  list of dataframes and compell to list for processing

```python
allproc = pd.concat(dfs)
print("There are {} processed Tweets".format(len(allproc)))
allproc= allproc.drop_duplicates()
allproc = allproc.sort_values(by=['Datetime']).reset_index(drop=True)
allproc['Datetime'] =  pd.to_datetime(allproc['Datetime'],infer_datetime_format=True)
allproc = allproc.drop(columns = ['isDate'])
print("After removing duplicates there are {} processed Tweets".format(len(allproc)))

allresult = pd.concat(dfs2)
print("There are {} processed Tweets".format(len(allresult)))
allresult= allresult.drop_duplicates()
allresult = allresult.sort_values(by=['Datetime']).reset_index(drop=True)
allresult['Datetime'] =  pd.to_datetime(allresult['Datetime'],infer_datetime_format=True)
print("After removing duplicates there are {} processed Tweets".format(len(allresult)))
```

```python
GeoTweets = pd.concat(dfs3)
GeoTweets = GeoTweets.sort_values(by=['Datetime']).reset_index(drop=True)
##there was an error parsing, looked like a tweet had a cairage return and got one level off in parsing
GeoTweets = GeoTweets.drop([10454])
GeoTweets = GeoTweets.astype({'target': 'float32'})
GeoTweets = GeoTweets.astype({'lat': 'float64'})
GeoTweets = GeoTweets.astype({'country_conf': 'float64'})
GeoTweets= GeoTweets.drop_duplicates()
GeoTweets['Datetime'] =  pd.to_datetime(GeoTweets['Datetime'],infer_datetime_format=True)
GeoTweets = GeoTweets.sort_values(by=['Datetime']).reset_index(drop=True)

GeoTweets['flood']=GeoTweets.ptext.str.contains('|'.join(floods))
GeoTweets['tropical']=GeoTweets.ptext.str.contains('|'.join(troplist))
GeoTweets['severeweather']=GeoTweets.ptext.str.contains('|'.join(wxevents))
GeoTweets['geomovement']=GeoTweets.ptext.str.contains('|'.join(movement))
GeoTweets['fire']=GeoTweets.ptext.str.contains('|'.join(flist))


alldfs = [allproc,allresult,GeoTweets]
dfname = ["All","Informative", "GeoTweets"]
```

```python
GeoTweets['Datetime'] =  pd.to_datetime(GeoTweets['Datetime'],infer_datetime_format=True)
```

```python
allproc_date_group = allproc.set_index('Datetime').groupby(pd.Grouper(freq='D')).count().assign(dataset='All Tweets')
allresult_date_group = allresult[allresult['target']==0].set_index('Datetime').groupby(pd.Grouper(freq='D')).count().assign(dataset='Informative Tweets')
GeoTweets_date_group = GeoTweets.set_index('Datetime').groupby(pd.Grouper(freq='D')).count().assign(dataset='GeoTweets')
```

```python
concatdates = pd.concat([allproc_date_group,allresult_date_group,GeoTweets_date_group] )
```

```python
for df in alldfs:
    df['day_of_week'] = df['Datetime'].dt.dayofweek
    df['hour']= df['Datetime'].dt.hour
    df = df
allproc = alldfs[0]
allresult = alldfs[1]
GeoTweets = alldfs[2]
```

```python
hour_dfs = []
day_of_week_dfs = []
for i in range(len(alldfs)):
    hourdf = alldfs[i].groupby('hour').count().assign(dataset=dfname[i])
    hour_dfs.append(hourdf)
    day_of_weekdf = alldfs[i].groupby('day_of_week').count().assign(dataset=dfname[i])
    day_of_week_dfs.append(day_of_weekdf)
    
hour = pd.concat(hour_dfs)
day_of_week = pd.concat(day_of_week_dfs)
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize =(10, 6))
ax = plt.subplot(projection='polar')

ax.set_theta_direction(-1)
ax.set_theta_zero_location("N")

t = hour.axes[0]
y = hour['TweetId']/61
tnorm = (t-t.min())/(t.max()-t.min())*2.*np.pi
# ax.fill_between(tnorm,y ,0, alpha=0.4)
ax.plot(tnorm,y , linewidth=0.8)
ax.set_title("1 June - 31 July Average Number of Tweets per Hour UTC",fontsize=14,fontweight='bold')
plt.thetagrids(range(0, 360, 15),(range(0,24,1)))
# plt.rgrids(range(0, max(test4['TweetId']), 5))

plt.show()
```
