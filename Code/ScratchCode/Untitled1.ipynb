```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Thu Jul 15 08:29:53 2021

@author: Arbo
"""
import os
import pandas as pd
code_dir = os.getcwd()
parent_dir = os.path.dirname(code_dir)
data_dir = os.path.join(parent_dir,"Data")
tweet_dir = os.path.join(parent_dir,"TweetMap")
def append_df(tweet_dir):
    
    tweets_df = pd.DataFrame(
                         columns=['Datetime', 'TweetId', 'Text',])
    
    for table in os.listdir(tweet_dir):
        if table.startswith("tweets"):
            tempdf = pd.read_csv(os.path.join(tweet_dir,table), engine = 'python', )
            
            tweets_df = tweets_df.append(tempdf)      
    tweets_df['Text'] = tweets_df['Text'].astype(str)        
    return tweets_df
```

```python
juneDFs=[]
for table in os.listdir(os.path.join(tweet_dir)):
    if table.startswith("geo"):
        tempdf = pd.read_csv(os.path.join(tweet_dir,table), engine = 'python', )
        juneDFs.append(tempdf)
```

```python

for x in range(len(juneDFs)):
    juneDFs[x] = juneDFs[x].drop(columns = ['TweetLink', '0'])

    print("Column Names for June Data{}: {} ".format(x,juneDFs[x].columns))
```

```python
JuneAll = pd.concat(juneDFs)
JuneAll= JuneAll.drop([4920,4921])
```

```python
JuneAll.to_csv(os.path.join(tweet_dir,"JuneTweets.csv"), index = False)
```

```python
julyDFs=[]
for table in os.listdir(os.path.join(tweet_dir,'tweets_archive')):
    if table.startswith("geo"):
        tempdf = pd.read_csv(os.path.join(tweet_dir,'tweets_archive', table), engine = 'python', )
        julyDFs.append(tempdf)
```

```python
for x in range(len(julyDFs)):
    try:
        julyDFs[x] = julyDFs[x].drop(columns = ['TweetLink', '0'])
    except:
        julyDFs[x] = julyDFs[x].drop(columns = ['0'])
    print("Column Names for July Data{}: {} ".format(x,julyDFs[x].columns))
```

```python
JulyAll = pd.concat(julyDFs)
```

```python
JulyAll=JulyAll.sort_values(by=['Datetime']).reset_index(drop=True)
```

```python
JulyAll.to_csv(os.path.join(tweet_dir,"JulyTweets.csv"), index = False)
```
